```python
import os
import pandas as pd

base_dir = "Driver Drowsiness Dataset (DDD)"

Drowsy = os.path.join(base_dir, "Drowsy")
Non_Drowsy = os.path.join(base_dir, "Non Drowsy")

filepaths = []
labels = []

class_dirs = [Drowsy, Non_Drowsy]
class_names = ["Drowsy", "Non_Drowsy"]

for i, folder in enumerate(class_dirs):
    for file in os.listdir(folder):
        fpath = os.path.join(folder, file)
        filepaths.append(fpath)
        labels.append(class_names[i])

bc_df = pd.DataFrame({
    "filepaths": filepaths,
    "labels": labels
})

print(bc_df["labels"].value_counts())
print(bc_df.shape)

bc_df.sample(10)
```

```python
from sklearn.model_selection import train_test_split

train_df, temp_df = train_test_split(
    bc_df,
    test_size=0.30,
    random_state=42,
    stratify=bc_df["labels"]
)

val_df, test_df = train_test_split(
    temp_df,
    test_size=0.50,
    random_state=42,
    stratify=temp_df["labels"]
)

print("Train set:", train_df.shape)
print("Validation set:", val_df.shape)
print("Test set:", test_df.shape)

print("\nTrain labels:\n", train_df["labels"].value_counts())
print("\nValidation labels:\n", val_df["labels"].value_counts())
print("\nTest labels:\n", test_df["labels"].value_counts())
```

```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split

dataset_path = "Driver Drowsiness Dataset (DDD)"

drowsy_path = os.path.join(dataset_path, "Drowsy")
non_drowsy_path = os.path.join(dataset_path, "Non Drowsy")

filepaths = []
labels = []

for img in os.listdir(drowsy_path):
    filepaths.append(os.path.join(drowsy_path, img))
    labels.append("Drowsy")

for img in os.listdir(non_drowsy_path):
    filepaths.append(os.path.join(non_drowsy_path, img))
    labels.append("Non-Drowsy")

bc_df = pd.DataFrame({
    "filepaths": filepaths,
    "labels": labels
})

train_df, temp_df = train_test_split(
    bc_df,
    test_size=0.30,
    random_state=42,
    stratify=bc_df["labels"]
)

val_df, test_df = train_test_split(
    temp_df,
    test_size=0.50,
    random_state=42,
    stratify=temp_df["labels"]
)

print("Train:", train_df.shape)
print("Val:", val_df.shape)
print("Test:", test_df.shape)

```

```python
import matplotlib.pyplot as plt
import numpy as np

def bc_images(image_gen, class_names):
    images, labels = next(image_gen)   
    plt.figure(figsize=(20, 20))

    r = min(len(labels), 25)

    for i in range(r):
        plt.subplot(5, 5, i + 1)

        img = images[i]
        img = (img - img.min()) / (img.max() - img.min())

        plt.imshow(img)
        label_index = np.argmax(labels[i])
        plt.title(class_names[label_index], color="green", fontsize=16)
        plt.axis('off')

    plt.show()
```

```python
import cv2
import mediapipe as mp

mp_face_mesh = mp.solutions.face_mesh
mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles

face_mesh = mp_face_mesh.FaceMesh(
    max_num_faces=1,
    refine_landmarks=True,
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5
)

cap = cv2.VideoCapture(0)
print("Camera started  press Q to quit")

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = face_mesh.process(rgb)

    if results.multi_face_landmarks:
        for landmarks in results.multi_face_landmarks:
            mp_drawing.draw_landmarks(
                frame,
                landmarks,
                mp_face_mesh.FACEMESH_TESSELATION,
                None,
                mp_drawing_styles.get_default_face_mesh_tesselation_style()
            )
            mp_drawing.draw_landmarks(
                frame,
                landmarks,
                mp_face_mesh.FACEMESH_CONTOURS,
                None,
                mp_drawing_styles.get_default_face_mesh_contours_style()
            )

    cv2.imshow("Driver Drowsiness  Face Mesh", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp
import numpy as np
import time
import winsound

# MediaPipe setup
mp_face_mesh = mp.solutions.face_mesh
mp_drawing = mp.solutions.drawing_utils

face_mesh = mp_face_mesh.FaceMesh(
    max_num_faces=1,
    refine_landmarks=True,
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5
)

LEFT_EYE = [33, 160, 158, 133, 153, 144]
RIGHT_EYE = [362, 385, 387, 263, 373, 380]

def eye_aspect_ratio(eye):
    A = np.linalg.norm(eye[1] - eye[5])
    B = np.linalg.norm(eye[2] - eye[4])
    C = np.linalg.norm(eye[0] - eye[3])
    return (A + B) / (2.0 * C)

EAR_THRESHOLD = 0.25
DROWSY_TIME = 2.0  

start_time = None
status = "ATTENTIVE"
last_beep_time = 0

cap = cv2.VideoCapture(0)
print("Camera started. Press Q to quit.")

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    h, w = frame.shape[:2]
    rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = face_mesh.process(rgb)

    status = "ATTENTIVE"

    if results.multi_face_landmarks:
        for face_landmarks in results.multi_face_landmarks:

            left_eye = np.array([
                [face_landmarks.landmark[i].x * w,
                 face_landmarks.landmark[i].y * h]
                for i in LEFT_EYE
            ])

            right_eye = np.array([
                [face_landmarks.landmark[i].x * w,
                 face_landmarks.landmark[i].y * h]
                for i in RIGHT_EYE
            ])

            ear = (eye_aspect_ratio(left_eye) + eye_aspect_ratio(right_eye)) / 2

            nose_x = face_landmarks.landmark[1].x * w
            nose_y = face_landmarks.landmark[1].y * h

            center_x_min = w * 0.35
            center_x_max = w * 0.65
            center_y_min = h * 0.30
            center_y_max = h * 0.70

            if ear < EAR_THRESHOLD:
                if start_time is None:
                    start_time = time.time()
                elif time.time() - start_time >= DROWSY_TIME:
                    status = "DROWSY"
            else:
                start_time = None

                if not (center_x_min < nose_x < center_x_max and
                        center_y_min < nose_y < center_y_max):
                    status = "NO FOCUS"

            mp_drawing.draw_landmarks(
                frame,
                face_landmarks,
                mp_face_mesh.FACEMESH_CONTOURS
            )

    if status in ["DROWSY", "NO FOCUS"]:
        if time.time() - last_beep_time > 1:
            winsound.Beep(2500, 700)
            last_beep_time = time.time()

    color = (0, 255, 0)
    if status == "DROWSY":
        color = (0, 0, 255)
    elif status == "NO FOCUS":
        color = (0, 165, 255)

    cv2.putText(frame, status, (30, 60),
                cv2.FONT_HERSHEY_SIMPLEX, 1.5, color, 3)

    cv2.imshow("Driver Monitoring System", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
